# Transport Layer Guide

This notebook demonstrates the Transport Layer features: qubit transmission via teleportation, fidelity tracking and resource accounting.

## Imports and Setup

```python
from quantumnet.topology import Network, Host
from quantumnet.utils import Logger

Logger.get_instance().activate()
```

```python
network = Network()
network.set_ready_topology("Grade", 3, 3)
network.draw()
```

## 1. Running the Transport Layer

`transportlayer.run_transport_layer(alice_id, bob_id, num_qubits)` executes the full transmission protocol:
1. Checks if Alice has enough qubits; creates more if needed.
2. Finds a valid route via the network layer.
3. Computes the route fidelity from EPR pairs.
4. Teleports each qubit and records the final fidelity.
5. Retries up to `config.protocol.transport_max_attempts` times.

```python
alice_id, bob_id = 0, 8
num_qubits = 5

result = network.transportlayer.run_transport_layer(alice_id, bob_id, num_qubits)
print(f"\nTransmission result: {result}")
print(f"Current timeslot: {network.clock.now}")
```

## 2. Inspecting Teleported Qubits

Each teleported qubit stores detailed information: source, destination, route, fidelities and timeslot.

```python
teleported = network.transportlayer.get_teleported_qubits()
for i, info in enumerate(teleported):
    print(f"Qubit {i}: route={info['route']}, "
          f"F_alice={info['fidelity_alice']:.4f}, "
          f"F_route={info['fidelity_route']:.4f}, "
          f"F_final={info['F_final']:.4f}, "
          f"timeslot={info['timeslot']}")
```

## 3. Average Transport Fidelity

Computes the average final fidelity across all teleported qubits.

```python
avg_fid = network.transportlayer.avg_fidelity_on_transportlayer()
print(f"\nAverage transport fidelity: {avg_fid:.6f}")
```

## 4. Resource Usage

```python
print(f"Qubits used (transport layer): {network.transportlayer.get_used_qubits()}")
print(f"Total used qubits (all layers): {network.get_total_used_qubits()}")
print(f"Total used EPRs (all layers): {network.get_total_used_eprs()}")
```
