# Network Layer Guide

This notebook demonstrates the Network Layer features: route finding and entanglement swapping.

## Imports and Setup

```python
from quantumnet.topology import Network, Host
from quantumnet.utils import Logger

Logger.get_instance().activate()
```

```python
network = Network()
network.set_ready_topology("Grade", 3, 3)
network.draw()
```

## 1. Finding a Valid Route

`networklayer.short_route_valid(alice_id, bob_id)` searches for the shortest path between two hosts where **every intermediate channel has at least one EPR pair available**. Returns `None` if no valid route exists.

```python
route = network.networklayer.short_route_valid(0, 8)
print(f"\nValid route from 0 to 8: {route}")
```

## 2. Entanglement Swapping

`networklayer.entanglement_swapping(alice_id, bob_id)` performs entanglement swapping across the entire route, creating a virtual EPR pair between Alice and Bob even if they are not directly connected.

For each pair of adjacent segments:
1. Takes the first EPR from each channel.
2. Computes the swapping success probability from fidelities.
3. If successful, creates a virtual EPR with the combined fidelity.
4. Removes the consumed EPRs.

```python
result = network.networklayer.entanglement_swapping(0, 8)
print(f"\nEntanglement swapping result: {result}")
print(f"Current timeslot: {network.clock.now}")
```

## 3. Used Routes and Average Route Size

The network layer tracks all routes used and can compute their average length (in hops).

```python
print(f"Routes used: {network.networklayer.routes_used}")
print(f"Average route size (hops): {network.networklayer.get_avg_size_routes():.2f}")
```

## 4. Resource Usage

```python
print(f"EPRs used (network layer): {network.networklayer.get_used_eprs()}")
print(f"Total EPRs used (all layers): {network.get_total_used_eprs()}")
print(f"Total qubits used (all layers): {network.get_total_used_qubits()}")
```
