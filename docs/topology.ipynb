# Topology Guide — Nodes, Edges and Hosts

This notebook demonstrates how to create, manipulate and inspect the network topology: hosts, nodes, edges and their attributes.

## Imports

```python
from quantumnet.topology import Network, Host
from quantumnet.utils import Logger
```

## 1. Building a Network Manually

You can create a `Network`, instantiate individual `Host` objects, define their connections, and add them one by one.

```python
network = Network()

# Create hosts
h0 = Host(0)
h1 = Host(1)
h2 = Host(2)

# Define connections (edges) between hosts
h0.add_connection(1)
h0.add_connection(2)
h1.add_connection(2)

# Add hosts to the network
network.add_host(h0)
network.add_host(h1)
network.add_host(h2)

# Draw the network
network.draw()
```

## 2. Accessing Nodes and Edges

Nodes and edges can be accessed directly through the `Network` object or via its underlying `graph` (a `networkx.Graph`).

```python
# Directly from the network
print("Nodes:", list(network.nodes))
print("Edges:", list(network.edges))

# Via the underlying NetworkX graph
print("Graph nodes:", list(network.graph.nodes))
print("Graph edges:", list(network.graph.edges))
```

## 3. Host Dictionary

The network stores hosts in a dictionary with the format `{host_id: Host}`. The key is always an integer.

```python
print(network.hosts)

# Retrieve a specific host
host = network.get_host(0)
print("Host ID:", host.host_id)
print("Host info:", host.info())
```

## 4. Node and Edge Attributes

The graph nodes and edges can carry arbitrary attributes stored as dictionaries. The graph does **not** hold host properties like memory — use the `Host` object for that.

```python
# Node attributes — initially empty
print("Node 0 attrs:", network.graph.nodes[0])

# Add custom attributes to a node
network.graph.nodes[0]['label'] = 'Alice'
network.graph.nodes[0]['color'] = 'blue'
print("Node 0 attrs:", network.graph.nodes[0])

# Access all node data at once
print("All nodes data:", list(network.graph.nodes.data()))
```

```python
# Edge attributes
print("Edge (0,1) attrs:", network.graph.edges[0, 1])

# Add custom attributes to an edge
network.graph.edges[0, 1]['weight'] = 1.5
print("Edge (0,1) attrs:", network.graph.edges[0, 1])

# Access all edge data at once
print("All edges data:", list(network.graph.edges.data()))
```

## 5. Ready-to-Use Topologies

`set_ready_topology` creates a standard topology with hosts, channels and EPR pairs already initialized.

Available topologies:
- `"Grade"` (Grid) — requires two arguments (rows, columns)
- `"Linha"` (Line) — requires one argument (number of nodes)
- `"Anel"` (Ring) — requires one argument (number of nodes)

```python
# Grid topology 3x3
net_grid = Network()
net_grid.set_ready_topology("Grade", 3, 3)
print("Grid nodes:", list(net_grid.nodes))
print("Grid edges:", list(net_grid.edges))
net_grid.draw()
```

```python
# Line topology with 5 nodes
net_line = Network()
net_line.set_ready_topology("Linha", 5)
print("Line nodes:", list(net_line.nodes))
net_line.draw()
```

```python
# Ring topology with 6 nodes
net_ring = Network()
net_ring.set_ready_topology("Anel", 6)
print("Ring nodes:", list(net_ring.nodes))
net_ring.draw()
```

## 6. Host Memory and Routing Table

Each host holds a qubit memory and a routing table.

```python
host0 = net_grid.get_host(0)
print("Memory size:", len(host0.memory))
print("Routing table:", host0.routing_table)
print("Full info:", host0.info())
```
