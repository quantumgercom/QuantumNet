# Controller Guide

This notebook demonstrates the Controller, which manages routing tables for all hosts in the network.

## Imports and Setup

```python
from quantumnet.topology import Network, Host
from quantumnet.control import Controller
from quantumnet.utils import Logger

Logger.get_instance().activate()
```

```python
network = Network()
network.set_ready_topology("Grade", 3, 3)
network.draw()
```

## 1. Creating the Controller

The `Controller` receives a reference to the network and computes shortest-path routing tables using NetworkX.

```python
controller = Controller(network)
```

## 2. Registering Routing Tables

`register_routing_tables()` computes the shortest path from each host to every other host and stores the result in each host's routing table.

```python
controller.register_routing_tables()

for host_id in network.hosts:
    rt = network.hosts[host_id].routing_table
    print(f"Host {host_id}: {rt}")
```

## 3. Creating a Routing Table for a Single Host

`create_routing_table(host_id)` returns the routing table for a specific node without registering it.

```python
rt = controller.create_routing_table(0)
print(f"Routing table for host 0:")
for dest, path in rt.items():
    print(f"  -> {dest}: {path}")
```

## 4. Route Announcements

The controller can announce routes to intermediate nodes or to Alice and Bob.

```python
route = [0, 3, 6, 7, 8]

print("Announcing to all route nodes:")
controller.announce_to_route_nodes(route)

print("\nAnnouncing to Alice and Bob:")
controller.announce_to_alice_and_bob(route)
```

## 5. Host Info After Routing

```python
for host_id in [0, 4, 8]:
    info = network.get_host(host_id).info()
    print(f"Host {host_id}: memory={info['memory']} qubits, routes={len(info['routing_table'])} entries")
```
